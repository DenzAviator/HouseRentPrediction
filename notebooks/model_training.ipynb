```python
#customized ranking for each ordinal variable
Area_Type_categories=["Super Area", "Carpet Area", "Built Area"]
City_categories=['Kolkata', 'Mumbai', 'Bangalore', 'Delhi', 'Chennai', 'Hyderabad']
Furnishing_Status_categories=['Unfurnished', 'Semi-Furnished', 'Furnished']
Tenant_Preferred_categories=['Bachelors/Family', 'Bachelors', 'Family']
Point_of_Contact_categories=['Contact Owner', 'Contact Agent', 'Contact Builder']
```

```python
from sklearn.impute import SimpleImputer ##for handling missing values
from sklearn.preprocessing import StandardScaler ## Handling Feature Scaling
from sklearn.preprocessing import OrdinalEncoder ##Ordinal Encoder
##pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline=Pipeline(
    
    steps=[
        
        ("imputer",SimpleImputer()),
        ("scaler",StandardScaler())
        
    ]
    
)
```

```python
cat_pipeline=Pipeline(
    
    steps=[
        
        ('imputer',SimpleImputer(strategy="most_frequent")),
        ('ordinalencoder',OrdinalEncoder(categories=[Area_Type_categories,City_categories,Furnishing_Status_categories,Tenant_Preferred_categories,Point_of_Contact_categories]))
        
    ]
    
)
```

```python
preprocessor=ColumnTransformer(
     [
        
        ('num_pipeline',num_pipeline,numerical_cols),
        ('cat_pipeline',cat_pipeline,categorical_cols)
     ]
)
```

```python
## train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=30)
```

```python
import numpy as np
def evaluate_model(true,predicted):
    mae = mean_absolute_error(true,predicted)
    mse = mean_squared_error(true,predicted)
    rmse = np.sqrt(mean_squared_error(true,predicted))
    r2_square = r2_score(true,predicted)
    return mae, rmse, r2_square
```
